### Load data

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
import numpy as np
np.set_printoptions(threshold=np.nan)
```

```python
fmri = np.load('data/fmri.npy')
labels = np.load('data/fmri_labels.npy')
```

```python
num_subjects = len(fmri)
num_samples, num_rois = np.shape(fmri[0])
```

### Compute the connectivity

```python
conn_mtx = np.zeros((num_subjects, num_rois, num_rois))
```

```python
for subj in range(num_subjects):
    print("Processing subject {0}".format(subj + 1))
    
    fmri_ts = fmri[subj]
    conn_mtx[subj, ...] = np.corrcoef(fmri_ts.T)
```

### Filter connectivity matrices

```python
thres_conn_mtx = np.zeros_like(conn_mtx)
```

```python
from dyfunconn.graphs import threshold_omst_global_cost_efficiency
```

```python
for subj in range(num_subjects):
    print("Processing subject {0}".format(subj + 1))
    
    subj_conn_mtx = np.abs(conn_mtx[subj])
    _, CIJtree, _, _, _, _ = threshold_omst_global_cost_efficiency(subj_conn_mtx)
    
    thres_conn_mtx[subj] = CIJtree
```

```python
np.save('data/fmri_thres_conn_mtx.npy', thres_conn_mtx)
```
